```python
print('Part 2: 2D Plot')
print('1. Create a line plot of sin(x) and cos(x + π/2) for -2π < x < 2π where x increases at intervals of π/4.')
print('i. Make the sin(x) graph red and make the cos(x+π/2) graph green, Put both lines onto the same plot')
      
import matplotlib.pyplot as plt
from pylab import *
import math 
import numpy as np

#x increases at intervals of π/4 i.e it generates 17 points, by changing 90 to 17 in the below linspace we get a new plot
x = np.linspace(-2*math.pi , 2*math.pi, 90)

plt.plot(x , np.sin(x), color = 'r' )
plt.plot(x, np.cos(x+math.pi/2), color = 'g')
plt.title('Line Plot')
plt.xlabel('X values')
plt.ylabel('Sin(x) and Cos(x+pi/2)')
plt.legend(['X', 'Sin(x) and Cos(x+pi/2)'])
plt.show()

print('ii. Using the same info as above, make a subplot with 2 different graphs- one graph for sin(x) and one graph for cos(x+π/2)')

subplot(1,2,1)
title('Sin(X) plot')
plot(x,np.sin(x), color = 'r')

subplot(1,2,2)
title('Cos(x+pi/2) plot')
plot(x, np.cos(x+math.pi/2), color = 'g')
```
